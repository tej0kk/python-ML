```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('DATA2.csv')
data
```

```python
plt.hist(data['Target'])
```

```python
#memisahkan antara data fitur dan data label
fitur = data.drop(['Target', 'no'], axis=1) #ini data feature, fitur, independent variables
label = data['Target']#ini data label
```

```python
#memilih algoritma model yang akan kita gunakan
from sklearn.neighbors import KNeighborsClassifier

#menentukan hyperparameter model
model_knn = KNeighborsClassifier(n_neighbors=5, 
                                 weights='distance',
                                 algorithm='auto',
                                 metric='euclidean')



#memerintahkan model untuk mempelajari data
model_knn.fit(fitur,label)
```
